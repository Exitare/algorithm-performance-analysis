# Wall runtime Errors json dump sorting

```python
import json
import pandas as pd
with open("wallruntime_jobs.json") as f:
    jj=json.load(f)
df=pd.DataFrame(jj)
```

```python
df["info"].unique()
```

```python
# split jobs by ids
jjs={}
for j in jj:
    if j["tool_id"] in jjs:
        jjs[j["tool_id"]].append(j)
    else:
        jjs[j["tool_id"]]=[j]
```

```python
# rename ids
correct_key={}
for k,v in jjs.items():
    if(len(k.split("/"))==1):
        key=k
    else:
        key="_".join(k.split("/")[2:])
    correct_key[key]=v
```

```python
for k,v in correct_key.items():
    f=open("walltime_dump/"+k+".json", 'w')
    f.write(json.dumps(v, indent=4, sort_keys=True))
    f.close()
```

```python
correct_key.keys()
```

```python
amounts={k:len(v) for k,v in correct_key.items()}
```

```python
import operator
toolids=sorted(amounts.items(), key=operator.itemgetter(1), reverse=True)
toolids=[i[0] for i in toolids]
print(toolids)
```

```python
import os
def getfiles(dirpath):
    a = [s for s in os.listdir(dirpath)
         if os.path.isfile(os.path.join(dirpath, s))]
    a.sort(key=lambda s: os.path.getmtime(os.path.join(dirpath, s)), reverse=True)
    return a
walltime=getfiles("walltime_csv")
memory=getfiles("memory_csv")
```
